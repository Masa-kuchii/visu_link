```python
# os.chdir('C:/Users/iryolabo/Desktop/可視化春大会/travel_vis')
os.chdir("/Users/masayuki/desktop/visu_link-master/travel_vis")

class tra_vis():
    
    def __init__(self, tra_vol):
        self.tra_vol = tra_vol
#         主観的旅行時間を格納
        self.df_1_100_sub=read_csv("link_car_mu("+str(tra_vol)+"_FP_100_1_10.000000)_g.csv")
        self.df_1_10_sub=read_csv("link_car_mu("+str(tra_vol)+"_FP_10_1_10.000000)_g.csv")
        self.df_1_sub=read_csv("link_car_mu("+str(tra_vol)+"_FP_1_1_10.000000)_g.csv")
        self.df_10_sub=read_csv("link_car_mu("+str(tra_vol)+"_FP_1_10_10.000000)_g.csv")
        self.df_100_sub=read_csv("link_car_mu("+str(tra_vol)+"_FP_1_100_10.000000)_g.csv")
#     　　真の旅行時間を格納
        self.df_1_100_tru=read_csv("link_true("+str(tra_vol)+"_FP_100_1_10.000000)_g.csv")
        self.df_1_10_tru=read_csv("link_true("+str(tra_vol)+"_FP_10_1_10.000000)_g.csv")
        self.df_1_tru=read_csv("link_true("+str(tra_vol)+"_FP_1_1_10.000000)_g.csv")
        self.df_10_tru=read_csv("link_true("+str(tra_vol)+"_FP_1_10_10.000000)_g.csv")
        self.df_100_tru=read_csv("link_true("+str(tra_vol)+"_FP_1_100_10.000000)_g.csv")
        #         もろもろの値
        self.size = 5
        self.colum_num = len(self.df_1_100_sub.columns)-1
#         リスト
        self.sub_list = [self.df_1_100_sub, self.df_1_10_sub, self.df_1_sub, self.df_10_sub, self.df_100_sub]
        self.tru_list = [self.df_1_100_tru, self.df_1_10_tru, self.df_1_tru, self.df_10_tru, self.df_100_tru]
        self.name_list = [1/100, 1/10, 1, 10, 100]
#         エラー処理
#         for i in range(self.size):
#             if (self.colum_num != )  

    def link_car_mu(self):
        print("全ドライバーの主観的旅行時間分布の平均")
        print("OD交通量"+str(self.tra_vol)+"台")
        fig = plt.figure(figsize=(20,100))    
        for j in range(self.colum_num):
            plt.subplot(20, 4, j+1)
            plt.plot(self.df_1_100_sub["time"] , self.df_1_100_sub[str(j)],label ="γ=1/100",linewidth=3, color=cm.rainbow(0/12.0), linestyle='solid')
            plt.plot(self.df_1_10_sub["time"] ,self.df_1_10_sub[str(j)],label ="γ=1/10", linewidth=3, color=cm.rainbow(2/12.0), linestyle='solid')
            plt.plot(self.df_1_sub["time"] ,self.df_1_sub[str(j)],label ="γ=1", linewidth=3,color=cm.rainbow(4/12.0), linestyle='solid')
            plt.plot(self.df_10_sub["time"] ,self.df_10_sub[str(j)],label ="γ=10", linewidth=3,color=cm.rainbow(6/12.0), linestyle='solid')
            plt.plot(self.df_100_sub["time"] ,self.df_100_sub[str(j)],label ="γ=100", linewidth=3,color=cm.rainbow(8/12.0), linestyle='solid')       
            plt.title("linkid_"+ str(j))
            plt.ylim(500, 4000)
            plt.xlabel("時間")
#             plt.xlabel("時間",fontproperties=fp)
            plt.ylabel("全ドライバーの主観的旅行時間分布の平均値")
#             plt.ylabel("全ドライバーの主観的旅行時間分布の平均値",fontproperties=fp)
            plt.grid()
            plt.subplots_adjust(wspace=0.3, hspace=0.4)
            if (j == 0):
                plt.legend(bbox_to_anchor=(0, 1.2), loc='lower left', borderaxespad=0, fontsize=18)   


    def link_true(self):
        print("10秒ごとに取得したリンクの最新旅行時間")
        print("OD交通量"+str(self.tra_vol)+"台")
        fig = plt.figure(figsize=(20,100))    
        for j in range(self.colum_num):
            plt.subplot(20, 4, j+1)
            plt.plot(self.df_1_100_tru["time"] , self.df_1_100_tru[str(j)],label ="γ=1/100",linewidth=3, color=cm.rainbow(0/12.0), linestyle='solid')
            plt.plot(self.df_1_10_tru["time"] ,self.df_1_10_tru[str(j)],label ="γ=1/10", linewidth=3, color=cm.rainbow(2/12.0), linestyle='solid')
            plt.plot(self.df_1_tru["time"] ,self.df_1_tru[str(j)],label ="γ=1", linewidth=3,color=cm.rainbow(4/12.0), linestyle='solid')
            plt.plot(self.df_10_tru["time"] ,self.df_10_tru[str(j)],label ="γ=10", linewidth=3,color=cm.rainbow(6/12.0), linestyle='solid')
            plt.plot(self.df_100_tru["time"] ,self.df_100_tru[str(j)],label ="γ=100", linewidth=3,color=cm.rainbow(8/12.0), linestyle='solid')       
            plt.title("linkid_"+ str(j))
            plt.ylim(500, 50000)
            plt.xlabel("時間")
#             plt.xlabel("時間",fontproperties=fp)
            plt.ylabel("10秒ごとに取得したリンクの最新旅行時間")
#             plt.ylabel("10秒ごとに取得したリンクの最新旅行時間",fontproperties=fp)
            plt.grid()
            plt.subplots_adjust(wspace=0.3, hspace=0.4)
            if (j == 0):
                plt.legend(bbox_to_anchor=(0, 1.2), loc='lower left', borderaxespad=0, fontsize=18)   

    def link_compare(self):
        print("10秒ごとに取得したリンクの最新旅行時間 vs 全ドライバーの主観的旅行時間分布の平均")
        print("OD交通量"+str(self.tra_vol)+"台")
        fig = plt.figure(figsize=(20,100))    
        for j in range(self.colum_num):
            plt.subplot(40, 4, 2*j+1)
            plt.plot(self.df_1_100_tru["time"] , self.df_1_100_tru[str(j)],label ="γ=1/100",linewidth=3, color=cm.rainbow(0/12.0), linestyle='solid')
            plt.plot(self.df_1_10_tru["time"] ,self.df_1_10_tru[str(j)],label ="γ=1/10", linewidth=3, color=cm.rainbow(2/12.0), linestyle='solid')
            plt.plot(self.df_1_tru["time"] ,self.df_1_tru[str(j)],label ="γ=1", linewidth=3,color=cm.rainbow(4/12.0), linestyle='solid')
            plt.plot(self.df_10_tru["time"] ,self.df_10_tru[str(j)],label ="γ=10", linewidth=3,color=cm.rainbow(6/12.0), linestyle='solid')
            plt.plot(self.df_100_tru["time"] ,self.df_100_tru[str(j)],label ="γ=100", linewidth=3,color=cm.rainbow(8/12.0), linestyle='solid')       
            plt.title("linkid_"+ str(j))
            plt.ylim(500, 50000)
            plt.xlabel("時間")
#             plt.xlabel("時間",fontproperties=fp)
            plt.ylabel("10秒ごとに取得したリンクの最新旅行時間")
#             plt.ylabel("10秒ごとに取得したリンクの最新旅行時間",fontproperties=fp)
            plt.grid()
            plt.subplots_adjust(wspace=0.4, hspace=0.6)
        
            if (j == 0):
                plt.legend(bbox_to_anchor=(0, 1.2), loc='lower left', borderaxespad=0, fontsize=18) 
            
            plt.subplot(40, 4, 2*j+2)
            plt.plot(self.df_1_100_sub["time"] , self.df_1_100_sub[str(j)],label ="γ=1/100",linewidth=3, color=cm.rainbow(0/12.0), linestyle='solid')
            plt.plot(self.df_1_10_sub["time"] ,self.df_1_10_sub[str(j)],label ="γ=1/10", linewidth=3, color=cm.rainbow(2/12.0), linestyle='solid')
            plt.plot(self.df_1_sub["time"] ,self.df_1_sub[str(j)],label ="γ=1", linewidth=3,color=cm.rainbow(4/12.0), linestyle='solid')
            plt.plot(self.df_10_sub["time"] ,self.df_10_sub[str(j)],label ="γ=10", linewidth=3,color=cm.rainbow(6/12.0), linestyle='solid')
            plt.plot(self.df_100_sub["time"] ,self.df_100_sub[str(j)],label ="γ=100", linewidth=3,color=cm.rainbow(8/12.0), linestyle='solid')       
            plt.title("linkid_"+ str(j))
            plt.ylim(500, 4000)
            plt.xlabel("時間")
#             plt.xlabel("時間",fontproperties=fp)
            plt.ylabel("全ドライバーの主観的旅行時間分布の平均値")
#             plt.ylabel("全ドライバーの主観的旅行時間分布の平均値",fontproperties=fp)
            plt.grid()
            plt.subplots_adjust(wspace=0.4, hspace=0.6)
            if (j == 0):
                plt.legend(bbox_to_anchor=(0, 1.2), loc='lower left', borderaxespad=0, fontsize=18)   

    def network_travel(self):
        print("10秒ごとに取得したネットワーク全体の旅行時間 vs 全リンクに対する全ドライバーの主観的旅行時間分布の平均")
        print("OD交通量"+ str(self.tra_vol)+"台")
        fig = plt.figure(figsize=(20,50)) 
        for j in range(self.size):            
            plt.subplot(5, 2, 2*j+1)
            plt.plot(self.tru_list[j]["time"] , self.tru_list[j].sum(axis=1)-self.tru_list[j]["time"],label =("γ="+str(self.name_list[j])),linewidth=3, color=cm.rainbow(0/12.0), linestyle='solid')
            plt.title("情報伝播速度γ="+ str(self.name_list[j])+"におけるネットワーク旅行時間")
            plt.xlabel("時間")
#             plt.xlabel("時間",fontproperties=fp)
            plt.ylabel("10秒ごとに取得したネットワークの最新旅行時間")
#             plt.ylabel("10秒ごとに取得したネットワークの最新旅行時間",fontproperties=fp)
            plt.grid()
    
    
            plt.subplot(5, 2, 2*j+2)
            plt.plot(self.sub_list[j]["time"] , self.sub_list[j].sum(axis=1)-self.sub_list[j]["time"],label =("γ="+str(self.name_list[j])),linewidth=3, color=cm.rainbow(0/12.0), linestyle='solid')
            plt.title("情報伝播速度γ="+ str(self.name_list[j])+"において，ネットワークに対する全ドライバーの主観的旅行時間分布の平均")
            plt.xlabel("時間")
#             plt.xlabel("時間",fontproperties=fp)
            plt.ylabel("ネットワークに対する全ドライバーの主観的旅行時間分布の平均")
#             plt.ylabel("10秒ごとに取得したネットワークの最新旅行時間",fontproperties=fp)
            plt.grid()


    
#     def gridlock link(self):
        
        
tra_15122 = tra_vis(15122)
tra_15122.network_travel()
```
